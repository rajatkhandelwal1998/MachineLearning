```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=10*np.random.rand(100)
x
```

```python
def sin_model(x,sigma=0.2):
    noise =sigma*np.random.randn(len(x))
    return np.sin(5*x)+np.sin(0.5*x)+noise
y=sin_model(x)#noisy signal
```

```python
plt.figure(figsize=(16,8))
plt.errorbar(x,y,0.1,fmt='o')
```

```python
xfit=np.linspace(0,10,1000)
rfr=RandomForestRegressor(100)
rfr.fit(x[:,None],y)
yfit=rfr.predict(xfit[:,None])#predicted from noise signal
ytrue=sin_model(xfit,0)#actual data without noise
```

```python
plt.figure(figsize=(16,8))
plt.errorbar(x,y,0.1,fmt='o')
plt.plot(xfit,yfit,'-r',alpha=0.5)
plt.plot(xfit,ytrue,'-b',alpha=0.5)
```

```python
rfr2=RandomForestRegressor(100)
rfr2.fit(xfit[:,None],ytrue)
ypred=rfr2.predict(xfit[:,None])
plt.figure(figsize=(16,8))
plt.errorbar(x,y,0.1,fmt='o')
plt.plot(xfit,ypred,'-r',alpha=0.5)
plt.plot(xfit,ytrue,'-b',alpha=0.5)
```
