```python
boston_df=DataFrame(boston.data)
boston_df.columns=boston.feature_names
boston_df
```

```python
boston_df['price']=boston.target
boston_df
```

```python
boston_df.head()
```

```python
sns.lmplot('RM','price',data=boston_df)
```

```python
X=boston_df.RM
X.head()
```

```python
Y=boston_df.price
Y.head()
```

```python
X=np.array([[value,1] for value in X])
X
```

```python
m,c=(np.linalg.lstsq(X,Y,rcond=None)[0])#to find m and c for mx+c
print(m,c)
```

```python
x=boston_df.RM
plt.plot(x,m*x+c,'r',label='Best Fit Line')
```

```python
plt.plot(x,boston_df.price,'o')
```

```python
plt.plot(x,m*x+c,'r',label='Best Fit Line')
plt.plot(x,boston_df.price,'o')
```

```python
plt.hist(boston_df.price,bins=50)
plt.xlabel('Prices in $1000s')
plt.ylabel('Number Of Houses')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=54)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
lreg=LinearRegression()
lreg.fit(X_train,Y_train)
test_predict=lreg.predict(X_test)
#test_predict
```

```python
pd.concat([DataFrame(np.ravel(Y_test),columns=['Test Data']),DataFrame(test_predict,columns=['Test Pred'])],axis=1).head()
```

```python
print('RMSE',np.sqrt(np.mean((Y_test-test_predict)**2)))
```
