```python
import numpy as np
from matplotlib import pyplot as plt
```

```python

x=np.arange(0,11)
y=2*x+5
plt.figure(figsize=(10,10))

plt.title('Line Plot')
plt.plot(x,y,'p',color='red',linestyle='dashed')#points shape=p,s,o(pentagon,sqare,round)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)#0.1 for every points
y=np.sin(x)
plt.figure(figsize=(10,10))

plt.title('Line Plot')
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
x=np.linspace(0,2*np.pi,100)#100 for points
y=np.sin(x)
plt.figure(figsize=(10,10))

plt.title('Line Plot')
plt.plot(x,0*x)
plt.plot(x,y)
plt.plot(np.arange(-1,1,0.1)+np.pi,np.arange(-1,1,0.1))
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.5)#for points
y=np.tan(x)
plt.figure(figsize=(10,10))

plt.title('Tan Plot')
#lt.plot(x,0*x)
plt.plot(x,y,'s',color='black',linestyle='dashed')
plt.plot(x,2*x+5)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
