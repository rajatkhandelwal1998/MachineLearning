```python
import numpy as np
from pandas import Series,DataFrame
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_boston
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
boston=load_boston()
```

```python
#print(boston.DESCR)
```

```python
boston_df=DataFrame(boston.data)
boston_df.columns=boston.feature_names
boston_df
```

```python
boston_df.head()
```

```python
x=np.ravel(boston_df.RM)
x
```

```python
y=np.ravel(boston.target)
y
```

```python
rfr=RandomForestRegressor(506)
rfr.fit(x[:,None],y)
ypred=rfr.predict(x[:,None])
```

```python
print('RMSE is %.2f'%np.sqrt(np.mean(y-ypred)**2))
```

```python
pd.concat([DataFrame(y,columns=['y']),DataFrame(ypred,columns=['ypred'])],axis=1).head()
```
